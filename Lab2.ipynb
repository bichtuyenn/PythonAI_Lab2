```python
#Bai 1
X = [0.349526784, 1.6974435, 5.384308891, 2.044150596, \
     4.578814506, 3.241690807, 2.535931731, 2.210580888, \
     3.397474351, 5.972933146, 5.114704101]
Y = [0.254020646, 0.790556868, -0.81239532, 1.012143475, \
     -0.904558188, -0.167456361, 0.482547054, 0.878514378, \
     -0.210093715, -0.128786937, -0.866501299]
import matplotlib.pyplot as plt
import numpy as np
def F_sin(x):
  return np.sin(x)

plt.scatter(X, Y, color="red")

X_new = np.linspace(0, 6, 100)
Y_new = F_sin(X_new)
plt.plot(X_new, Y_new, color="green")

plt.title("Question 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#Bai 2
lr = 0.000008
a = 0
b = 0
c = 0
d = 0

for e in range(200_000):
  grad_a = 0
  grad_b = 0
  grad_c = 0
  grad_d = 0
  loss = 0
  
  for i in range(len(X)):
    x = X[i]
    y = Y[i]
    grad_a = grad_a - 2*x**3*(y - a*x**3 - b*x**2 - c*x - d)
    grad_b = grad_b - 2*x**2*(y - a*x**3 - b*x**2 - c*x - d)
    grad_c = grad_c - 2*x*(y - a*x**3 - b*x**2 - c*x - d)
    grad_d = grad_d - 2*(y - a*x**3 - b*x**2 - c*x - d)
    loss = loss + (y - a*x**3 - b*x**2 - c*x - d)**2

  a = a - lr*grad_a
  b = b - lr*grad_b
  c = c - lr*grad_c
  d = d - lr*grad_d

def f(x):
  return  a*x**3 + b*x**2 + c*x + d

plt.scatter(X, Y, color='red', marker='.')
X_2 = np.linspace(0, 6, 100)
Y_2 = f(X_2)
plt.plot(X_2, Y_2, color='blue')

X_new = np.linspace(0, 6, 100)
Y_new = F_sin(X_new)
plt.plot(X_new, Y_new, color='green')
plt.title('Question 2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Bai 3a
def F_sin(x):
  return np.sin(x)

plt.scatter(X, Y, color="red")

X_new = np.linspace(0,6,100)
Y_new = F_sin(X_new)
plt.plot(X_new, Y_new, color="green")

X3=[]
for x in X:
    X3 = X3 + [[1, x, x**2, x**3]]
X3 = np.array(X3)
Y3 = np.array([Y]).T
Z1 = np.matmul(X3.T, X3)
Z2 = np.linalg.inv(Z1)
Z3 = np.matmul(Z2, X3.T)
w = np.matmul(Z3, Y3)
print(w)

def F(x):
  return w[3][0]*x**3 + w[2][0]*x**2 + w[1][0]*x + w[0][0]

X_new = np.linspace(0, 6, 100)
Y_new = F(X_new)
plt.plot(X_new, Y_new, color="blue")

plt.title("Question 3a")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
